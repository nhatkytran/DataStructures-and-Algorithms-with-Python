```python
class Trie:
    def __init__(self):
        self.links = [None] * 26
        self.ew = 0
        self.cp = 0
    
    def index(self, char):
        return ord(char) - ord('a')
    
    def contains(self, char):
        return self.links[self.index(char)] is not None
    
    def push(self, char):
        self.links[self.index(char)] = self.__class__()

    def get(self, char):
        return self.links[self.index(char)]

    def increaseEndWith(self):
        self.ew += 1

    def increasePrefix(self):
        self.cp += 1

    def decreaseEndWith(self):
        self.ew -= 1

    def decreasePrefix(self):
        self.cp -= 1

    def getEndWith(self):
        return self.ew
    
    def getPrefix(self):
        return self.cp


class Tries:
    def __init__(self):
        self.root = Trie()

    def insert(self, word):
        cur = self.root
        for char in word:
            if not cur.contains(char):
                cur.push(char)
            cur = cur.get(char)
            cur.increasePrefix()
        cur.increaseEndWith()

    def countWordsEndWith(self, word):
        cur = self.root
        for char in word:
            if not cur.contains(char):
                return 0
            cur = cur.get(char)
        return self.getEndWith()
    
    def countWordsStartWith(self, word):
        cur = self.root
        for char in word:
            if not cur.contains(char):
                return 0
            cur = cur.get(char)
        return cur.getPrefix()
    
    # Assume that erase only word that exists in Tries
    def erase(self, word):
        cur = self.root
        for char in word:
            cur = cur.get(char)
            cur.decreasePrefix()
        cur.decreaseEndWith()


tries = Tries()

tries.insert('apple')
tries.insert('apple')
tries.insert('apps')
tries.insert('apps')
```
