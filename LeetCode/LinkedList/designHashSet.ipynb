```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class MyHashSet:
    def __init__(self):
        self.head = None

    def add(self, key: int) -> None:
        if not self.head:
            self.head = Node(key)
        else:
            current = self.head
            while current.next:
                if current.value == key:
                    return
                current = current.next
            if current.value != key:
                current.next = Node(key)

    def remove(self, key: int) -> None:
        if self.head:
            prev = None
            current = self.head
            while current:
                if current.value == key:
                    if not prev:
                        self.head = self.head.next
                    else:
                        prev.next = current.next
                    return
                prev = current
                current = current.next

    def contains(self, key: int) -> bool:
        current = self.head
        while current:
            if current.value == key:
                return True
            current = current.next
        return False
    

class MyHashSet:
    def __init__(self, size=50):
        self.table = [None] * size
        self.size = size

    def hash(self, key):
        return sum(ord(item) for item in str(key)) % self.size
    
    def add(self, key):
        hash_value = self.hash(key)
        if not self.table[hash_value]:
            self.table[hash_value] = [key]
        elif key not in self.table[hash_value]:
            self.table[hash_value].append(key)

    def contains(self, key):
        hash_value = self.hash(key)
        if not self.table[hash_value]:
            return False
        return key in self.table[hash_value]
    
    def remove(self, key):
        hash_value = self.hash(key)
        if not self.table[hash_value]: return
        self.table[hash_value] = [value for value in self.table[hash_value] \
                                    if value != key]
        if not self.table[hash_value]:
            self.table[hash_value] = None
```
