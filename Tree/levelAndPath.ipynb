```python
class NodeTree:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


node0 = NodeTree(0)
node1 = NodeTree(1)
node2 = NodeTree(2)
node3 = NodeTree(3)
node4 = NodeTree(4)
node5 = NodeTree(5)

node0.left = node1
node0.right = node2
node1.left = node3
node1.right = node4
node2.right = node5


# Binary Tree


#       0
#     /   \
#    1     2
#  /  \      \
# 3    4       5


def adjacency_iterative(root):
    step = 0
    ad_list = dict()
    ad_list[step] = [root]
    step = 1 if root.left or root.right else None
    while step:
        go_next = False
        for node in ad_list[step - 1]:
            left = node.left
            right = node.right
            try:
                ad_list[step]
            except KeyError:
                ad_list[step] = []
            finally:
                nodes = []
                if left:
                    ad_list[step].append(left)
                    nodes.append(left)
                if right:
                    ad_list[step].append(right)
                    nodes.append(right)
                for node in nodes:
                    go_next = True if node.left or node.right else go_next
        step = step + 1 if go_next else None
    return ad_list


def adjacency_recursive(root):
    ad_list = dict()
    def inner(root, step=0):
        if not root:
            return
        try:
            ad_list[step]
        except KeyError:
            ad_list[step] = []
        finally:
            ad_list[step].append(root)
            inner(root.left, step + 1)
            inner(root.right, step + 1)
    inner(root)
    return ad_list


# Adjacency


# {
#     0: [node0],
#     1: [node1, node2],
#     2: [node3, node4, node5]
# }


ad_list = adjacency_iterative(node0)
ad_list = adjacency_recursive(node0)


def check_level_iterative(ad_list, value):
    for key, nodes in ad_list.items():
        for node in nodes:
            if node.value == value:
                return key


def check_level_recursive(ad_list, value):
    keys = list(ad_list.keys())
    def inner(index=0):
        try:
            key = keys[index]
            for node in ad_list[key]:
                if node.value == value:
                    return key
            return inner(index + 1)
        except IndexError:
            return
    return inner()


def path_iterative(ad_list, value):
    level = check_level_iterative(ad_list, value)
    if level == 0:
        [root] = ad_list[level]
        return [root.value]
    path = [value]
    current_value = value
    current_level = level - 1
    while current_level >= 0:
        current_node = None
        for node in ad_list[current_level]:
            left = node.left
            right = node.right
            if left and left.value == current_value \
                    or right.value == current_value:
                current_node = node
                break
        path = [current_node.value, *path]
        current_value = current_node.value
        current_level -= 1
    return path


def path_recursive(ad_list, value):
    path = [value]
    level = check_level_recursive(ad_list, value)
    def inner(current_level, current_value=value):
        if current_level < 0:
            return path
        current_node = None
        for node in ad_list[current_level]:
            left = node.left
            right = node.right
            if left and left.value == current_value \
                    or right and right.value == current_value:
                current_node = node
                break
        path.insert(0, current_node.value)
        return inner(current_level - 1, current_node.value)
    return inner(level - 1)


def path_recursive2(ad_list, value):
    level = check_level_recursive(ad_list, value)
    def inner(current_level, current_value=value):
        if current_level < 0:
            return []
        current_node = None
        for node in ad_list[current_level]:
            left = node.left
            right = node.right
            if left and left.value == current_value \
                    or right and right.value == current_value:
                current_node = node
                break
        value = current_node.value
        return [*inner(current_level - 1, value), value]
    return [*inner(level - 1), value]
```
