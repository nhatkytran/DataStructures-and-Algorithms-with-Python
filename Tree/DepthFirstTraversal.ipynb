```python
class Stack:
    def __init__(self):
        self.top = None
        self.size = 0

    def __len__(self):
        return self.size

    class NodeStack:
        def __init__(self, value):
            self.value = value
            self.next = None

    def __repr__(self):
        represent = ''
        current = self.top
        while current:
            represent += f'{current.value} --> '
            current = current.next
        return represent

    def push(self, value):
        new_node = self.NodeStack(value)
        if not self.top:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.size += 1

    def pop(self):
        if not self.size:
            return None
        popped_node = self.top
        self.top = self.top.next
        self.size -= 1
        return popped_node.value


#       0
#     /   \
#    1     2
#  /  \      \
# 3    4       5


class NodeTree:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


node0 = NodeTree(0)
node1 = NodeTree(1)
node2 = NodeTree(2)
node3 = NodeTree(3)
node4 = NodeTree(4)
node5 = NodeTree(5)

node0.left = node1
node0.right = node2
node1.left = node3
node1.right = node4
node2.right = node5


def depthFirstTraversal(root):
    if not root:
        return
    stack = Stack()
    stack.push(root)
    while stack.size:
        popped_node = stack.pop()
        print(popped_node.value)
        if popped_node.right:
            stack.push(popped_node.right)
        if popped_node.left:
            stack.push(popped_node.left)


def get_tree_size(root):
    if not root:
        return
    stack = Stack()
    stack.push(root)
    size = 0
    while stack.size:
        popped_node = stack.pop()
        size += 1
        if popped_node.right:
            stack.push(popped_node.right)
        if popped_node.left:
            stack.push(popped_node.left)
    return size


def get_tree_total(root):
    if not root:
        return
    stack = Stack()
    stack.push(root)
    total = 0
    while stack.size:
        popped_node = stack.pop()
        total += popped_node.value
        if popped_node.right:
            stack.push(popped_node.right)
        if popped_node.left:
            stack.push(popped_node.left)
    return total


def recursiveDFT(current):
    if not current:
        return
    print(current.value)
    recursiveDFT(current.left)
    recursiveDFT(current.right)


def recursiveDFT_size(current):
    return 1 + recursiveDFT_size(current.left) + recursiveDFT_size(current.right) \
        if current else 0


def recursiveDFT_total(current):
    return current.value + recursiveDFT_size(current.left) + recursiveDFT_size(current.right) \
        if current else 0
```
