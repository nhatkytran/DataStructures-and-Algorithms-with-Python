```python
class Stack:
    def __init__(self):
        self.top = None
        self.size = 0

    def __len__(self):
        return self.size

    class NodeStack:
        def __init__(self, value):
            self.value = value
            self.next = None

    def __repr__(self):
        represent = ''
        current = self.top
        while current:
            represent += f'{current.value} --> '
            current = current.next
        return represent

    def push(self, value):
        new_node = self.NodeStack(value)
        if not self.size:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.size += 1

    def pop(self):
        if not self.size:
            return None
        popped_node = self.top
        self.top = self.top.next
        self.size -= 1
        return popped_node.value


class NodeTree:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


node0 = NodeTree(0)
node1 = NodeTree(1)
node2 = NodeTree(2)
node3 = NodeTree(3)
node4 = NodeTree(4)
node5 = NodeTree(5)

node0.left = node1
node0.right = node2
node1.left = node3
node1.right = node4
node2.right = node5


#      0
#    /   \
#   1     2
#  /  \    \
# 3    4    5


# Pre-order: Self, Left, Right

def pre_order(self):
    if not self:
        return
    print(self.value)
    pre_order(self.left)
    pre_order(self.right)


# pre_order(node0)
# 0 1 3 4 2 5


# Post-order: Left, Right, Self

def post_order(self):
    if not self:
        return
    post_order(self.left)
    post_order(self.right)
    print(self.value)


# post_order(node0)
# 3 4 1 5 2 0


# In-order: Left, Self, Right

def in_order(self):
    if not self:
        return
    in_order(self.left)
    print(self.value)
    in_order(self.right)


# in_order(node0)
# 3 1 4 0 2 5
```
