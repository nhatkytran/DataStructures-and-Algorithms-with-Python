```python
class Queue:
    def __init__(self):
        self.top = None
        self.end = None
        self.size = 0

    def __len__(self):
        return self.size

    class NodeQueue:
        def __init__(self, value):
            self.value = value
            self.next = None

    def __repr__(self):
        represent = ''
        current = self.top
        while current:
            represent += f'{current.value} --> '
            current = current.next
        return represent

    def push(self, value):
        new_node = self.NodeQueue(value)
        if not self.size:
            self.top = new_node
            self.end = new_node
        else:
            self.end.next = new_node
            self.end = self.end.next
        self.size += 1

    def pop(self):
        popped_node = None
        if not self.size:
            return popped_node
        elif self.size == 1:
            popped_node = self.top
            self.top = None
            self.end = None
        else:
            popped_node = self.top
            self.top = self.top.next
        self.size -= 1
        return popped_node.value

    def includes(self, value):
        current = self.top
        while current:
            if current.value == value:
                return True
            current = current.next
        return False


# A --> C
# |     |
# \/    \/
# B     E
# |
# \/
# D --> F


graph = {
    'a': ['b', 'c'],
    'b': ['d'],
    'c': ['e'],
    'd': ['f'],
    'e': [],
    'f': []
}


def breath_first(graph, source):
    queue = Queue()
    queue.push(source)
    while queue.size:
        popped = queue.pop()
        print(popped)
        for neighbor in graph[popped]:
            queue.push(neighbor)


graph = {
    'f': ['g', 'i'],
    'g': ['h'],
    'h': [],
    'i': ['g', 'k'],
    'j': ['i'],
    'k': []
}


def has_path(graph, src, dst):
    queue = Queue()
    queue.push(src)
    while queue.size:
        popped = queue.pop()
        print(popped)
        if popped == dst:
            return True
        for neighbor in graph[popped]:
            queue.push(neighbor)
    return False


def has_path_bk(graph, src, dst):
    queue = Queue()
    queue_bk = Queue()
    queue.push(src)
    while queue.size:
        popped = queue.pop()
        if not queue_bk.includes(popped):
            if popped == dst:
                return True
            for neighbor in graph[src]:
                queue.push(neighbor)
            queue_bk.push(popped)
    return False
```
