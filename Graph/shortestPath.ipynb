```python
class Queue:
    def __init__(self):
        self.top = None
        self.end = None
        self.size = 0

    def __len__(self):
        return self.size

    class NodeQueue:
        def __init__(self, value):
            self.value = value
            self.next = None

    def __repr__(self):
        represent = ''
        current = self.top
        while current:
            represent += f'{current.value} --> '
            current = current.next
        return represent

    def push(self, value):
        new_node = self.NodeQueue(value)
        if not self.size:
            self.top = new_node
            self.end = new_node
        else:
            self.end.next = new_node
            self.end = self.end.next
        self.size += 1

    def pop(self):
        popped_node = None
        if not self.size:
            return popped_node
        elif self.size == 1:
            popped_node = self.top
            self.top = None
            self.end = None
        else:
            popped_node = self.top
            self.top = self.top.next
        self.size -= 1
        return popped_node.value

    def __contains__(self, value):
        current = self.top
        while current:
            if current.value == value:
                return True
            current = current.next
        return False


edges = [
  ['w', 'x'],
  ['x', 'y'],
  ['z', 'y'],
  ['z', 'v'],
  ['w', 'v'],
]


def build_graph(edges):
    graph = dict()
    for edge in edges:
        for current_node in edge:
            new_nodes = [node for node in edge if node != current_node]
            try:
                graph[current_node] = [*graph[current_node], *new_nodes]
            except KeyError:
                graph[current_node] = new_nodes
    return graph


graph = build_graph(edges)


graph = {
    'w': ['x', 'v'],
    'x': ['w', 'y'],
    'y': ['x', 'z'],
    'z': ['y', 'v'],
    'v': ['z', 'w']
}


def shortest_path(graph, src, dst):
    Node = namedtuple('Node', ('src', 'step'))
    queue = Queue()
    queue_bk = Queue()
    queue.push(Node(src, 0))
    queue_bk.push(src)
    while queue.size:
        popped = queue.pop()
        if popped.src == dst:
            return popped.step
        for neighbor in graph[popped.src]:
            if neighbor not in queue_bk:
                queue_bk.push(neighbor)
                queue.push(Node(neighbor, popped.step + 1))
    return -1

```
