```python
# Point from start
def start(list_, target):
    for i in range(len(list_)):
        if list_[i] > target:
            return i
    return -1


def recursive_start(list_, target, pointer=0):
    if pointer == len(list_):
        return -1
    if list_[pointer] > target:
        return pointer
    return recursive_start(list_, target, pointer + 1)


def recursive_start_affl(list_, target, pointer=0):
    try:
        if list_[pointer] > target:
            return pointer
        return recursive_start_affl(list_, target, pointer + 1)
    except IndexError:
        return -1


# Point from end
def end(list_, target):
    result = -1
    for i in range(len(list_)):
        last = len(list_) - (i + 1)
        if list_[last] > target:
            result = last
            continue
        break
    return result


def recursive_end(list_, target, result=-1):
    if not len(list_):
        return result
    if list_[-1] > target:
        result = len(list_) - 1
        return recursive_end(list_[:-1], target, result)
    return result


def recursive_end_affl(list_, target, result=-1):
    try:
        if list_[-1] > target:
            result = len(list_) - 1
            return recursive_end(list_[:-1], target, result)
        return result
    except IndexError:
        return result


# Point from center
def center(list_, target):
    pointer = len(list_) // 2
    result = -1
    if list_[pointer] <= target:
        for i in range(pointer + 1, len(list_)):
            if list_[i] > target:
                result = i
                break
        return result
    if list_[pointer] > target:
        result = pointer
        for i in range(pointer):
            new_pointer = pointer - (i + 1)
            if list_[new_pointer] > target:
                result = new_pointer
                continue
            break
        return result


def recursive_center(list_, target, pointer=None, result=-1, found=False):
    if pointer is None:
        pointer = len(list_) // 2
    if pointer < 0 or pointer == len(list_):
        return result
    if list_[pointer] <= target:
        if found:
            return result
        return recursive_center(list_, target, pointer + 1, result, found)
    if list_[pointer] > target:
        result = pointer
        found = True
        return recursive_center(list_, target, pointer - 1, result, found)


# Two points
def two(list_, target):
    result = -1
    for i in range(len(list_)):
        left = i
        right = len(list_) - (i + 1)
        if list_[left] > target:
            result = left
            break
        if list_[right] <= target:
            break
        if list_[right] > target:
            result = right
    return result


def recursive_two(list_, target, pointer=0, result=-1):
    left = pointer
    right = len(list_) - (pointer + 1)
    if list_[left] > target:
        return left
    if list_[right] <= target:
        return result
    result = right
    return recursive_two(list_, target, pointer + 1, result)
```
