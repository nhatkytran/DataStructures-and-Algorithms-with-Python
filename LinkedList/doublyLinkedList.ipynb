```python
class LinkedList:
    def __init__(self):
        self.head = None

    class Node:
        def __init__(self, value):
            self.prev = None
            self.value = value
            self.next = None

    def __len__(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

    def __repr__(self):
        if not self.head:
            return ''
        represent = '<-- '
        current = self.head
        while current:
            represent += f'{current.value} <--> ' \
                    if current.next else f'{current.value} --> '
            current = current.next
        return represent

    def append(self, value):
        new_node = self.Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def contains(self, value):
        current = self.head
        while current:
            if current.value == value:
                return True
            current = current.next
        return False

    def sum(self):
        total = 0
        current = self.head
        while current:
            total += current.value
            current = current.next
        return total

    def remove_index(self, index):
        count = 0
        is_removed = False
        current = self.head
        while current:
            if count == index:
                if current is self.head:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                else:
                    current.prev.next = current.next
                    if current.next:
                        current.next.prev = current.prev
                is_removed = True
                break
            count += 1
            current = current.next
        print('Remove successfully') if is_removed \
            else print('Index not found!')

    def remove_index_reverse(self, index):
        is_removed = False
        current = self.head
        while current:
            if index == 0:
                if current is self.head:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                else:
                    current.prev.next = current.next
                    if current.next:
                        current.next.prev = current.prev
                is_removed = True
                break
            index -= 1
            current = current.next
        print('Remove successfully') if is_removed \
            else print('Index not found!')

    def remove_value(self, value):
        is_removed = False
        current = self.head
        while current:
            if current.value == value:
                if current is self.head:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                else:
                    current.prev.next = current.next
                    if current.next:
                        current.next.prev = current.prev
                is_removed = True
                break
            current = current.next
        print('Remove successfully') if is_removed \
            else print('Value not found!')

    def find_reverse(self, index):
        current = self.head
        while current:
            if index == 0:
                return current.value
            index -= 1
            current = current.next
        return None

    def insert(self, value, index):
        is_inserted = False
        count = 0
        current = self.head
        new_node = self.Node(value)
        while current:
            if count == index:
                if current is self.head:
                    new_node.next = self.head
                    self.head.prev = new_node
                    self.head = new_node
                else:
                    current.prev.next = new_node
                    new_node.prev = current.prev
                    current.prev = new_node
                    new_node.next = current
                is_inserted = True
                break
            count += 1
            current = current.next
        print('Insert successfully') if is_inserted \
            else print('Index not found!')

    def insert_reverse(self, value, index):
        is_inserted = False
        current = self.head
        new_node = self.Node(value)
        while current:
            if index == 0:
                if current is self.head:
                    new_node.next = self.head
                    self.head.prev = new_node
                    self.head = new_node
                else:
                    current.prev.next = new_node
                    new_node.prev = current.prev
                    current.prev = new_node
                    new_node.next = current
                is_inserted = True
                break
            index -= 1
            current = current.next
        print('Insert successfully') if is_inserted \
            else print('Index not found!')

    def zip(self, list2):
        if not isinstance(list2, self.__class__):
            raise ValueError('Wrong type!')
        if not self.head or not list2.head:
            raise AttributeError('Empty linked list!')
        cur1, cur2 = self.head, list2.head
        while cur1.next and cur2.next:
            next1, next2 = cur1.next, cur2.next
            cur1.next = cur2
            cur2.prev = cur1
            cur2.next = next1
            next1.prev = cur2
            cur1, cur2 = next1, next2
        if not cur1.next:
            cur1.next = cur2
            cur2.prev = cur1
        else:
            next1 = cur1.next
            cur1.next = cur2
            cur2.prev = cur1
            cur2.next = next1
            next1.prev = cur2

    def check_even(self, number):
        return not bool(number % 2)

    def zip_even(self, list2):
        if not isinstance(list2, self.__class__):
            raise ValueError('Wrong type!')
        if not self.head or not list2.head:
            raise AttributeError('Empty linked list!')
        count = 0
        tail = self.head
        cur1 = self.head.next
        cur2 = list2.head
        while cur1 and cur2:
            if self.check_even(count):
                tail.next = cur2
                cur2.prev = tail
                cur2 = cur2.next
            else:
                tail.next = cur1
                cur1.prev = tail
                cur1 = cur1.next
            count += 1
            tail = tail.next
        if not cur1:
            tail.next = cur2
            cur2.prev = tail
        else:
            tail.next = cur1
            cur1.prev = tail
    

linked_list = LinkedList()
```
