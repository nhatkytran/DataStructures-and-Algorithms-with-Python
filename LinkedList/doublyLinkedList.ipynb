```python
class LinkedList:
    def __init__(self):
        self.head = None

    class Node:
        def __init__(self, value):
            self.prev = None
            self.value = value
            self.next = None

    def __len__(self):
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

    def __repr__(self):
        if not self.head:
            return ''
        represent = '<-- '
        current = self.head
        while current:
            represent += f'{current.value} <--> ' \
                    if current.next else f'{current.value} --> '
            current = current.next
        return represent

    def append(self, value):
        new_node = self.Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current

    def contains(self, value):
        current = self.head
        while current:
            if current.value == value:
                return True
            current = current.next
        return False

    def sum(self):
        total = 0
        current = self.head
        while current:
            total += current.value
            current = current.next
        return total

    def remove_index(self, index):
        count = 0
        is_removed = False
        current = self.head
        while current:
            if count == index:
                if current is self.head:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                else:
                    current.prev.next = current.next
                    if current.next:
                        current.next.prev = current.prev
                is_removed = True
                break
            count += 1
            current = current.next
        print('Remove successfully') if is_removed \
            else print('Index not found!')

    def remove_index_reverse(self, index):
        is_removed = False
        current = self.head
        while current:
            if index == 0:
                if current is self.head:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                else:
                    current.prev.next = current.next
                    if current.next:
                        current.next.prev = current.prev
                is_removed = True
                break
            index -= 1
            current = current.next
        print('Remove successfully') if is_removed \
            else print('Index not found!')

    def remove_value(self, value):
        is_removed = False
        current = self.head
        while current:
            if current.value == value:
                if current is self.head:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                else:
                    current.prev.next = current.next
                    if current.next:
                        current.next.prev = current.prev
                is_removed = True
                break
            current = current.next
        print('Remove successfully') if is_removed \
            else print('Value not found!')

    def find_reverse(self, index):
        current = self.head
        while current:
            if index == 0:
                return current.value
            index -= 1
            current = current.next
        return None


linked_list = LinkedList()

for i in range(3):
    linked_list.append(i)

print(linked_list)
print(linked_list)
```
