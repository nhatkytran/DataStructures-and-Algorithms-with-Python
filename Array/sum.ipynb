```python
# Point from start
def start(list_):
    result = 0
    for number in list_:
        result += number
    return result


def recursive_start(list_):
    if not len(list_):
        return 0
    return list_[0] + recursive_start(list_[1:])


def recursive_start_v2(list_, result=0):
    if not len(list_):
        return result
    result += list_[0]
    return recursive_start_v2(list_[1:], result)


# Two points
def two(list_):
    length = len(list_)
    result = 0
    for left in range(length):
        right = length - (left + 1)
        if left > right:
            return result
        if left == right:
            result += list_[left]
        if left < right:
            result += list_[left] + list_[right]
    return result


def recursive_two(list_, left=None, right=None):
    if left is None or right is None:
        left = 0
        right = len(list_) - 1
    if left > right:
        return 0
    if left == right:
        return list_[left]
    if left < right:
        return list_[left] + list_[right] + recursive_two(list_, left + 1, right - 1)


def recursive_two_v2(list_, left=None, right=None, result=0):
    if left is None or right is None:
        left = 0
        right = len(list_) - 1
    if left > right:
        return result
    if left == right:
        return result + list_[left]
    if left < right:
        result = list_[left] + list_[right]
        return recursive_two(list_, left + 1, right - 1, result)


# Recursive refactor for point from start
def refactor(list_):
    def inner(list_, pointer=0):
        if pointer == len(list_):
            return 0
        return list_[pointer] + inner(list_, pointer + 1)
    return inner(list_)


def refactor_v2(list_):
    def inner(list_, pointer=0, result=0):
        if pointer == len(list_):
            return result
        result += list_[pointer]
        return inner(list_, pointer + 1, result)
    return inner(list_)
```
