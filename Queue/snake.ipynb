```python
class Queue:
    def __init__(self):
        self.top = None
        self.end = None
        self.size = 0
    
    class NodeQueue:
        def __init__(self, value):
            self.value = value
            self.next = None

    def __len__(self):
        return self.size

    def __repr__(self):
        represent = ''
        current = self.top
        while current:
            represent += f'{current.value} --> '
            current = current.next
        return represent

    def __iter__(self):
        return self.Iterator(self.top)

    class Iterator:
        def __init__(self, current):
            self.current = current

        def __iter__(self):
            return self

        def __next__(self):
            if not self.current:
                raise StopIteration
            result = self.current.value
            self.current = self.current.next
            return result

    def push(self, value):
        new_node = self.NodeQueue(value)
        if not self.size:
            self.top = new_node
            self.end = new_node
        else:
            self.end.next = new_node
            self.end = self.end.next
        self.size += 1

    def pop(self):
        popped_node = None
        if not self.size:
            return popped_node
        elif self.size == 1:
            popped_node = self.top
            self.top = None
            self.end = None
        else:
            popped_node = self.top
            self.top = self.top.next
        return popped_node


class Snake:
    def __init__(self):
        self.snake = Queue()
        [self.snake.push([4, i + 1]) for i in range(4)]
        self.dimension = 10

    def move(self, direction):
        delta = { 'up': [-1, 0], 'down': [1, 0], 'left': [0, -1], 'right': [0, 1] }
        [currentRow, currentColumn] = self.snake.end.value
        [changeRow, changeColumn] = delta[direction]
        new_head = [currentRow + changeRow, currentColumn + changeColumn]
        self.snake.pop()
        self.snake.push(new_head)
        print('---')
        self.draw()
        print('---')

    def draw(self):
        grid = []
        for _ in range(self.dimension):
            row = []
            for j in range(self.dimension):
                row.append(' ')
            grid.append(row)
        for pos in self.snake:
            [row, col] = pos
            grid[row][col] = 'o'
        for row in grid:
            print('|'.join(row))

            
snake = Snake()
```
