```python
class Solution:
    def numUniqueEmails(self, emails):
        for i in range(len(emails)):
            [local_name, domain_name] = emails[i].split('@')
            local_name = local_name.split('+')[0]
            local_name = ''.join(local_name.split('.'))
            emails[i] = f'{local_name}@{domain_name}'
        return len(set(emails))
    
    def numUniqueEmails(self, emails):
        hash_set = set()
        for email in emails:
            [local_name, domain_name] = email.split('@')
            local_name = local_name.split('+')[0]
            local_name = ''.join(local_name.split('.'))
            hash_set.add(f'{local_name}@{domain_name}')
        return len(hash_set)
    
    def numUniqueEmails(self, emails):
        hash_set = set()
        for email in emails:
            i = 0
            local_name = ''
            domain_name = ''
            while email[i] not in ['@', '+']:
                if email[i] != '.':
                    local_name += email[i]
                i += 1
            while email[i] != '@':
                i += 1
            for i in range(i + 1, len(email)):
                domain_name += email[i]
            hash_set.add(f'{local_name}@{domain_name}')
        return len(hash_set)
    
    def numUniqueEmails(self, emails):
        hash_set = set()
        for email in emails:
            local_name = ''
            for i in range(len(email)):
                if email[i] in ['@', '+']:
                    break
                if email[i] != '.':
                    local_name += email[i]
            domain_name = ''
            for i in range(len(email) - 1, -1, -1):
                if email[i] == '@':
                    break
                domain_name = f'{email[i]}{domain_name}'
            hash_set.add(f'{local_name}@{domain_name}')
        return len(hash_set)
    
    def numUniqueEmails(self, emails):
        hash_set = set()
        for email in emails:
            i, local_name, i_stop = 0, '', False
            j, domain_name, j_stop = len(email) - 1, '', False
            while i != j:
                if i_stop and j_stop:
                    break
                if email[i] not in ['@', '+']:
                    if email[i] != '.':
                        local_name += email[i]
                    i += 1
                else:
                    i_stop = True

                if email[j] != '@':
                    domain_name = f'{email[j]}{domain_name}'
                    j -= 1
                else:
                    j_stop = True
            hash_set.add(f'{local_name}@{domain_name}')
        return len(hash_set)


s = Solution()

# Time complexity --> O(n * m) --> m --> Do some stuffs like split, join
# Space complexity --> O(n)


print(s.numUniqueEmails(["test.email+alex@leetcode.com", "test.e.mail+bob.cathy@leetcode.com", "testemail+david@lee.tcode.com"]))
print(s.numUniqueEmails(["a@leetcode.com", "b@leetcode.com", "c@leetcode.com"]))
print(s.numUniqueEmails(["test.email+alex@leetcode.com", "test.email.leet+alex@code.com"]))
```
