```python
def partition(nums, left, right):
    pivot = nums[right]
    start = left
    for i in range(left, right):
        if nums[i] < pivot:
            nums[start], nums[i] = nums[i], nums[start]
            start += 1
    nums[start], nums[right] = pivot, nums[start]
    return start

def quick_sort(nums):
    def inner(left, right):
        if left < right:
            pivot_index = partition(nums, left, right)
            inner(left, pivot_index - 1)
            inner(pivot_index + 1, right)
    inner(0, len(nums) - 1)
    return nums


class Solution:
    def sortedSquares(self, nums):
        for i in range(len(nums)):
            nums[i] = nums[i] ** 2
        quick_sort(nums)
        return nums
    
    def sortedSquares(self, nums):
        left = 0
        right = len(nums) - 1
        point = None
        while left <= right:
            middle = left + (right - left) // 2
            if nums[middle] >= 0:
                point = middle
                right = middle - 1
            else:
                left = middle + 1
        for i in range(len(nums)):
            nums[i] = nums[i] ** 2
        if point is None:
            point = len(nums) - 1
        i = point
        j = point - 1
        result = []
        while i < len(nums) and j >= 0:
            if nums[i] <= nums[j]:
                result.append(nums[i])
                i += 1
            else:
                result.append(nums[j])
                j -= 1
        while i < len(nums):
            result.append(nums[i])
            i += 1
        while j >= 0:
            result.append(nums[j])
            j -= 1
        return result
    
    def sortedSquares(self, nums):
        left = 0
        right = len(nums) - 1
        result = [None] * len(nums)
        while left <= right:
            if nums[right] ** 2 >= nums[left] ** 2:
                result[right - left] = nums[right] ** 2
                right -= 1
            else:
                result[right - left] = nums[left] ** 2
                left += 1
        return result


s = Solution()

print(s.sortedSquares([-4, -1, 0, 3, 10]))
print(s.sortedSquares([-7, -3, 2, 3, 11]))
print(s.sortedSquares([-5, -3, -2, -1]))
```
