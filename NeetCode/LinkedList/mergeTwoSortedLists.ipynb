```python
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

    class Node:
        def __init__(self, value):
            self.value = value
            self.next = None

    def view(self):
        rpr = ''
        current = self.head
        while current:
            rpr += f'{current.value} --> '
            current = current.next
        print(rpr)

    def append(self, value):
        node = self.Node(value)
        if not self.head:
            self.head = node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = node
        self.size += 1

    def extend(self, iterable):
        for item in iterable:
            self.append(item)

    @staticmethod
    def mergeTwoLists(list1, list2):
        if not list1.head: return list2.head
        if not list2.head: return list1.head
        head = None
        if list1.head.value <= list2.head.value:
            head = list1.head
            current = list1.head
            cur1 = current.next
            cur2 = list2.head
        else:
            head = list2.head
            current = list2.head
            cur1 = list1.head
            cur2 = current.next
        while cur1 and cur2:
            if cur1.value <= cur2.value:
                current.next = cur1
                current = cur1
                cur1 = cur1.next
            else:
                current.next = cur2
                current = cur2
                cur2 = cur2.next
        current.next = cur1 if cur1 else cur2
        return head
    

ll1 = LinkedList()
ll2 = LinkedList()

ll1.extend([1, 2, 4])
ll2.extend([1, 3, 4])

ll1.view()
ll2.view()

head = LinkedList.mergeTwoLists(ll1, ll2)

current = head
while current:
    print(current.value)
    current = current.next
```
