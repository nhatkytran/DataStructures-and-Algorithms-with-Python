```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def __repr__(self):
        rpr = ''
        current_node = self.head
        while current_node:
            rpr += f'{current_node.value} --> '
            current_node = current_node.next
        return rpr

    def add_node(self, value):
        if self.head is None:
            self.head = Node(value)
        else:
            current_node = self.head
            while current_node.next is not None:
                current_node = current_node.next
            current_node.next = Node(value)


ll1 = LinkedList()
for num in [1, 2, 2, 1]:
    ll1.add_node(num)

ll2 = LinkedList()
for num in [1, 2]:
    ll2.add_node(num)


class Solution:
    def isPalindrome(self, head):
        values = []
        while head:
            values.append(head.value)
            head = head.next
        return values == values[::-1]
    
    def isPalindrome(self, head):
        values = []
        current = head
        while current:
            values.append(current.value)
            current = current.next
        current = head
        for i in range(len(values) - 1, -1, -1):
            if values[i] != current.value:
                return False
            current = current.next
        return True
    
    def isPalindrome(self, head):
        # Find the middle of Linked List
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        # Reverse half of the linked list --> start from middle
        prev = slow
        slow = slow.next
        prev.next = None
        while slow:
            next_node = slow.next
            slow.next = prev
            prev = slow
            slow = next_node
        # 1 --> 2 --> 3 <-- 2 <-- 1
        # 1 --> 2 --> 2 <-- 1
        start = head
        end = prev
        while end:
            if start.value != end.value:
                return False
            start = start.next
            end = end.next
        return True
    
    def isPalindrome(self, head):
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        prev = None
        while slow:
            next_node = slow.next
            slow.next = prev
            prev = slow
            slow = next_node
        left = head
        right = prev
        while right:
            if left.value != right.value:
                return False
            left = left.next
            right = right.next
        return True


s = Solution()

print(s.isPalindrome(ll1.head))
print(s.isPalindrome(ll2.head))
```
