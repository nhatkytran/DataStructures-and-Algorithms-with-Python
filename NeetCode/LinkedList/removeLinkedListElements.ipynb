```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def __repr__(self):
        rpr = ''
        current_node = self.head
        while current_node:
            rpr += f'{current_node.value} --> '
            current_node = current_node.next
        return rpr

    def add_node(self, value):
        if self.head is None:
            self.head = Node(value)
        else:
            current_node = self.head
            while current_node.next is not None:
                current_node = current_node.next
            current_node.next = Node(value)


ll = LinkedList()
for num in [1, 2, 6, 3, 4, 5, 6]:
    ll.add_node(num)

print(ll)


class Solution:
    def removeElements(self, head, val):
        new_head = head
        prev = None
        while head:
            if head.value == val:
                if not prev:
                    head = head.next
                    new_head = head
                else:
                    prev.next = head.next
                    head = head.next
            else:
                prev = head
                head = head.next
        return new_head
    
    def removeElements(self, head, val):
        if not head:
            return None
        if head.value == val:
            return self.removeElements(head.next, val)
        head.next = self.removeElements(head.next, val)
        return head

    def removeElements(self, head, val):
        if not head:
            return None
        head.next = self.removeElements(head.next, val)
        return head.next if head.value == val else head
    
    def removeElements(self, head, val):
        if not head:
            return None
        next_node = self.removeElements(head.next, val)
        if head.value == val:
            return next_node
        head.next = next_node
        return head
    
    def removeElements(self, head, val):
        smart = Node(-1)
        smart.next = head
        head = smart
        while head.next:
            if head.next.value == val:
                head.next = head.next.next
            else:
                head = head.next
        return smart.next


s = Solution()

ll.head = s.removeElements(ll.head, 6)

print(ll)
```
