```python
import string


class Solution:
    def isPalindrome(self, s):
        new_s = ''
        for letter in s:
            letter = letter.lower()
            if letter in string.ascii_lowercase or letter in string.digits:
                new_s += letter
        left = 0
        right = len(new_s) - 1
        while left < right:
            if new_s[left] != new_s[right]:
                return False
            left += 1
            right -= 1
        return True
    
    def isPalindrome(self, s):
        left = 0
        right = len(s) - 1
        valid_letters = string.ascii_lowercase + string.digits
        while left < right:
            left_letter = s[left].lower()
            right_letter = s[right].lower()
            if left_letter not in valid_letters:
                left += 1
                continue
            if right_letter not in valid_letters:
                right -= 1
                continue
            if left_letter != right_letter:
                return False
            left += 1
            right -= 1
        return True
    
    def isPalindrome(self, s):
        letters = []
        for letter in s:
            if 48 <= ord(letter) <= 57 or 65 <= ord(letter) <= 90 or 97 <= ord(letter) <= 122:
                letters.append(letter)
        left = 0
        right = len(letters) - 1
        while left < right:
            if letters[left].lower() != letters[right].lower():
                return False
            left += 1
            right -= 1
        return True
    
    def isPalindrome(self, s: str) -> bool:
        l, r = 0, len(s) - 1
        while l < r:
            while l < r and not self.alphanum(s[l]):
                l += 1
            while l < r and not self.alphanum(s[r]):
                r -= 1
            if s[l].lower() != s[r].lower():
                return False
            l += 1
            r -= 1
        return True
    
    def alphanum(self, c):
        return (
            ord("A") <= ord(c) <= ord("Z")
            or ord("a") <= ord(c) <= ord("z")
            or ord("0") <= ord(c) <= ord("9")
        )
    
    def isPalindrome(self, s):
        new_string = ''
        for letter in s:
            if letter.isalnum():
                new_string += letter.lower()
        return new_string == new_string[::-1]


s = Solution()

print(s.isPalindrome("A man, a plan, a canal: Panama"))
print(s.isPalindrome("race a car"))
print(s.isPalindrome(" "))
```
