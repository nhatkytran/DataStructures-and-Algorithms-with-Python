```python
class BST:
    def __init__(self):
        self.root = None

    class Node:
        def __init__(self, value):
            self.value = value
            self.left = None
            self.right = None

    def see(self):
        if not self.root:
            raise ValueError('No root found!')
        queue = [self.root]
        while queue:
            popped = queue.pop(0)
            print(popped.value)
            if popped.left: queue.append(popped.left)
            if popped.right: queue.append(popped.right)

    def insert(self, value):
        if self.insert_travel(value, self.root):
            self.root = self.Node(value)

    def insert_travel(self, value, current):
        if not current:
            return True
        if value == current.value:
            raise ValueError('Value already exists!')
        if value < current.value:
            if self.insert_travel(value, current.left):
                current.left = self.Node(value)
        if value > current.value:
            if self.insert_travel(value, current.right):
                current.right = self.Node(value)

    def delete(self, value):
        self.root = self.delete_travel(value, self.root)

    def delete_travel(self, value, current):
        if not current:
            return None
        if value == current.value:
            if not current.left and not current.right:
                return None
            elif current.left and not current.right:
                return current.left
            elif current.right and not current.left:
                return current.right
            else:
                point = current.right
                while point.left: point = point.left
                current.value = point.value
                current.right = self.delete_travel(point.value, current.right)
        if value < current.value:
            current.left = self.delete_travel(value, current.left)
        if value > current.value:
            current.right = self.delete_travel(value, current.right)
        return current

    def find_node(self, value):
        def inner(current):
            if not current:
                return None
            if value == current.value:
                return current
            if value < current.value:
                return inner(current.left)
            if value > current.value:
                return inner(current.right)
        return inner(self.root)

    def find_min_node(self, current):
        while current.left:
            current = current.left
        return current

    def successor(self, value):
        current = self.find_node(value)
        if not current:
            return None
        if current.right:
            return self.find_min_node(current.right).value
        ancestor = self.root
        _successor = None
        while ancestor is not current:
            if current.value > ancestor.value:
                ancestor = ancestor.right
            if current.value < ancestor.value:
                _successor = ancestor.value
                ancestor = ancestor.left
        return _successor

    def inorder(self):
        cache = []
        def inner(current):
            if not current:
                return
            inner(current.left)
            cache.append(current.value)
            inner(current.right)
        inner(self.root)
        return cache

    def inorder_successor(self, value):
        numbers = self.inorder()
        for i in range(len(numbers) - 1):
            if numbers[i] == value:
                return numbers[i + 1]

    def precessor(self, value):
        return self.precessor_travel(value, self.root)

    def precessor_travel(self, value, current):
        if not current:
            return None
        if value == current.value:
            if current.left:
                point = current.left
                while point.right:
                    point = point.right
                return point.value
            else:
                ancestor = self.root
                _precessor = None
                while ancestor is not current:
                    if current.value < ancestor.value:
                        ancestor = ancestor.left
                    if current.value > ancestor.value:
                        _precessor = ancestor.value
                        ancestor = ancestor.right
                return _precessor
        if value < current.value:
            return self.precessor_travel(value, current.left)
        if value > current.value:
            return self.precessor_travel(value, current.right)

    def inorder_precessor(self, value):
        numbers = self.inorder()
        for i in range(1, len(numbers)):
            if numbers[i] == value:
                return numbers[i - 1]

    def successor_iterative(self, value):
        successor = None
        current = self.root
        while current:
            if value >= current.value:
                current = current.right
            elif value < current.value:
                successor = current.value
                current = current.left
        return successor


bst = BST()

numbers = [15, 10, 20, 8, 9, 12, 17, 25]
for number in numbers:
    bst.insert(number)
```
